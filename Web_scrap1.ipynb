```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://collider.com/movies-with-the-most-imdb-votes/#39-django-unchained-39-2012"
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_movies = soup.find_all('span', class_="item-num")
```

```python
scraped_movies
```

```python
scraped_movies1 = soup.find_all('h2')
```

```python
scraped_movies1
```

```python
movies_num = []
movies_name = []
for movie_span in scraped_movies1:
    movie = movie_span.text.strip()
    spl_movie = movie.split("  ")
    movies_num.append(spl_movie[0])
    movies_name.append(spl_movie[1])

print(movies_num)
print(movies_name)
    
```

```python
data = pd.DataFrame()
data["numbers"] = movies_num
data["movies"] = movies_name
movies_list = data.iloc[::-1]
movies_list
```

```python
movies_list.to_csv('movies_list.csv', index=False)
```
