```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.chrome.service import Service
```

```python
path = 'C:\\chromedriver.exe'
```

```python
service = Service(executable_path = path)
browser = webdriver.Chrome(service=service)
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
```

```python
browser.get(url)
```

```python
country_list = browser.find_elements("xpath", "//h3[@class='country-name']")
```

```python
country_list
```

```python
countries = []
for country in country_list:
    countries.append(country.text)
countries
```

```python
capital_list = browser.find_elements("xpath","//span[@class='country-capital']")
```

```python
capital_list
```

```python
capitals = []
for capital in capital_list:
    capitals.append(capital.text)
capitals
```

```python
population_list = browser.find_elements("xpath", "//span[@class='country-population']")
populations = []
for population in population_list:
    populations.append(population.text)
populations
```

```python
area_list = browser.find_elements("xpath", "//span[@class='country-area']")
areas = []
for area in area_list:
    areas.append(area.text)
areas
```

```python
data1 = pd.DataFrame()
data1["country"] = countries
data1["capital"] = capitals
data1["population"] = populations
data1["area"] = areas
data1
```

```python
data1.to_csv("countries_data.csv", index=False)
```
