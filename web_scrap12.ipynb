```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.chrome.service import Service
from time import sleep
```

```python
path = "C://chromedriver.exe"
```

```python
service = Service(executable_path = path)
browser = webdriver.Chrome(service = service)
```

```python
browser.get("https://www.amazon.in/")
```

```python
browser.maximize_window()
```

```python
input_search = browser.find_element("id",'twotabsearchtextbox')
search_button = browser.find_element("xpath","//input[@type='submit']")
```

```python
input_search.send_keys("Smartphones under 10000")
sleep(1)
search_button.click()
```

```python
products = [] 
for i in range(10):
    print("scrapping page:",i+1)
    product = browser.find_elements("xpath", "//span[@class='a-size-medium a-color-base a-text-normal']")
    for p in product:
        products.append(p.text)
    next_button = browser.find_element("xpath", "//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    next_button.click()
    sleep(3)
```

```python
products
```

```python
mobile = []
for product in products:
    temp = product.split("|")[0]
    ptemp = temp.split("(")[0]
    print(ptemp)
    mobile.append(ptemp)
```

```python
data_phones = pd.DataFrame()
data_phones["Mobile"] = mobile
```

```python
data_phones.to_csv("Mobile_under_10000.csv", index=False)
```
